```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import colorsys
%matplotlib inline
```

```python
df = pd.read_csv('/Users/xingobar/Downloads/2016-FCC-New-Coders-Survey-Data.csv')
```

```python
df.head()
```

# Age

```python
fig,ax = plt.subplots(figsize=(10,10))
plt.title('Age distribution',fontsize=15)
plt.hist(df['Age'].dropna(),bins=15,color = "#ffcc80",alpha=0.7,label='Age')
plt.ylabel('Frequency',fontsize=15)
plt.xlabel('Age',fontsize=15)
plt.setp(ax.get_xticklabels(),fontsize=15)
plt.setp(ax.get_yticklabels(),fontsize=15)
plt.legend(loc='best',fontsize='x-large')
plt.show()
```

# Gender population

```python
labels = df['Gender'].value_counts().index
fig,ax = plt.subplots(figsize=(10,10))
colors = ['#f4511e','#00acc1','#4caf50','#fb8c00','#9575cd']
patches, texts, autotexts = plt.pie(df['Gender'].value_counts(), autopct='%1.1f%%', 
                                    shadow=True, startangle=90,colors=colors)

### fontsize
for idx in xrange(len(autotexts)):
    autotexts[idx].set_fontsize(15)

plt.legend(labels,fontsize='x-large',loc='best', bbox_to_anchor=(1.05,1),frameon=True)
plt.title('Gender',fontsize=15)
plt.show()
```

# Employment Field

```python
employment_field = df['EmploymentField'].value_counts().index
employment_counts = df['EmploymentField'].value_counts().values
```

```python
fig,ax = plt.subplots(figsize=(10,10))
length = len(df.EmploymentField.value_counts().index)
HSV_tuples = [(x*1.0/length, 0.8, 0.8) for x in xrange(length)]
RGB = [colorsys.hsv_to_rgb(*i) for i in HSV_tuples]
patches, texts, autotexts = plt.pie(df['EmploymentField'].value_counts(), 
                                    autopct='%1.1f%%', shadow=True, startangle=90,colors=RGB)
### fontsize
for text in xrange(len(autotexts)):
    autotexts[text].set_fontsize(15)

#autotexts[0].set_fontsize(15)
plt.legend(employment_field,fontsize='x-large',loc='best', bbox_to_anchor=(1.05,1),frameon=True)
plt.title('EmploymentField',fontsize=15)
plt.show()
```

# Age Group

```python
df['AgeGroup'] = pd.cut(df['Age'],[0,21,31,41,99],labels=['10-20','21-30','31-40','41+'],right=False)
```

```python
fig,ax = plt.subplots(figsize=(10,10))
df['AgeGroup'].value_counts(sort=False).plot(kind='bar',ax=ax,rot=0)
plt.title('Age Distribution',fontsize=15)
plt.ylabel('Number of coder' , fontsize=15)
plt.xlabel('Age',fontsize=15)
ticks = plt.setp(ax.get_xticklabels(),fontsize=15)
ticks = plt.setp(ax.get_yticklabels(),fontsize=15)
```

# Month Programming

```python
df['MonthsProgrammingGroup'] = pd.cut(df['MonthsProgramming'],[0,6,24,999],labels=['<6','6-24','24+'],right=False)
fig,ax = plt.subplots(figsize=(10,10))
df['MonthsProgrammingGroup'].value_counts(sort=False).plot(kind='bar',ax=ax,rot=0)
plt.title('Month Programming',fontsize=15)
plt.xlabel('Month Group',fontsize=15)
plt.ylabel('Number of coders',fontsize=15)
ticks = plt.setp(ax.get_xticklabels(),fontsize=15)
ticks = plt.setp(ax.get_yticklabels(),fontsize=15)
```

# Hours Learning

```python
df['HoursLearningGroup'] = pd.cut(df['HoursLearning'],[0,6,13,24,999],labels=['<6','6-12','13-24','24+'])
fig,ax = plt.subplots(figsize=(10,10))
plt.title('Hours Learning',fontsize=15)
df['HoursLearningGroup'].value_counts(sort=False).plot(kind='bar',ax=ax,rot=0)
plt.ylabel('Number of coders',fontsize=15)
plt.xlabel('Hour Group',fontsize=15)
ticks = plt.setp(ax.get_xticklabels(),fontsize=15)
ticks = plt.setp(ax.get_yticklabels(),fontsize=15)
```

# MoneyForLearning

```python
df['MoneyForLearningGroup'] = pd.cut(df['MoneyForLearning'],[0,500,1000,200000],labels=['<500','500-1000','1000+'])
fig,ax = plt.subplots(figsize=(10,5))
plt.title('MoneyForLearning',fontsize=15)
df['MoneyForLearningGroup'].value_counts(sort=False).plot(kind='bar',ax=ax,rot=0)
plt.ylabel('Number of coders',fontsize=15)
plt.xlabel('Money',fontsize=15)
ticks = plt.setp(ax.get_xticklabels(),fontsize=15)
ticks = plt.setp(ax.get_yticklabels(),fontsize=15)
#print df['MoneyForLearning'].unique()
```
