```python
# https://www.kaggle.com/anokas/2015-us-traffic-fatalities
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
accidents = pd.read_csv('/Users/xingobar/Downloads/ACC_AUX.CSV')
```

```python
state  = accidents.STATE.value_counts().sort_index()
FATALS = accidents.FATALS.value_counts().sort_index()
state_type = accidents.STATE.unique()
FATALS_type = accidents.FATALS.unique()
```

```python
plt.figure(figsize=(18,10))
fig,ax = plt.subplots(1,1)
for i in range(1,11):
    accidents['A_REGION'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('A_REGION')
plt.ylabel('Total accident')
plt.title('The total accident of region')
```

```python
accidents.A_REGION.value_counts()
```

```python
plt.figure(figsize=(18,10))
fig,ax = plt.subplots(1,1)
accidents['FATALS'].value_counts().sort_index().plot(kind='barh')
ax.set_ylabel('FATALS type')
ax.set_xlabel('Total')
plt.title('The total of FATALS within types')
```
