```python
lats = []
lons= []
magnitudes = []
timestrings = []
for i in xrange(len(df)):
    lats.append(df['latitude'][i])
    lons.append(df['longitude'][i])
    magnitudes.append(df['mag'][i])
    timestrings.append(df['time'][i])

print 'longitude : ' ,lons[0:5]
print 'latitude : ' ,lats[0:5]
```

```python
## building map
map =Basemap(projection='robin', resolution = 'l', area_thresh = 1000.0,
              lat_0=0, lon_0=-130)

map.drawmapboundary()
map.drawcountries()
map.fillcontinents(color = 'gray')
map.drawcoastlines()
map.drawmeridians(np.arange(0, 360, 30))
map.drawparallels(np.arange(-90, 90, 30))

def get_marker_color(magnitude):
    # Returns green for small earthquakes, yellow for moderate
    #  earthquakes, and red for significant earthquakes.
    if magnitude < 3.0:
        return ('go')
    elif magnitude < 5.0:
        return ('yo')
    else:
        return ('ro')




min_marker_size = 2.5
for lon,lat,mag in zip(lons,lats,magnitudes):
    x,y = map(lon,lat)
    msize = min_marker_size * mag
    marker_string = get_marker_color(msize)
    map.plot(x,y,marker_string,markersize=msize)

title_string = 'Earthquake of Magnitude 1.0 or Greater'
title_string += "%s through %s" % (timestrings[-1][:10], timestrings[0][:10])
plt.title(title_string)
plt.show()
```
