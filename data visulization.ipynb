```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
df.describe()
```

```python
df.drop(['Ticket','Cabin'],axis=1,inplace=True)
```

```python
df
```

```python
# drop NA
df.dropna(inplace=True)
```

# Create graphics

```python
fig = plt.figure(figsize=(18,6),dpi=600)
alpha_barchart = 0.55
alpha = 0.2

## figure 1 ##
ax1 = plt.subplot2grid((2,3),(0,0))  # two row , three column 
df['Survived'].value_counts().plot(kind='bar',alpha=alpha_barchart)
ax1.set_xlim(-1,2)
plt.title('Distribution of Survived ( 1=Survived)')

### figure 2 ##
ax2 = plt.subplot2grid((2,3),(0,1))
plt.scatter(df['Survived'],df['Age'],alpha=alpha)
plt.title('Scatter of Survived by Age (1=Survived) ')
plt.ylabel('Age')
plt.grid(b=True,which='major',axis='y')

## figure 3 ##
ax3 = plt.subplot2grid((2,3),(0,2))
df['Pclass'].value_counts().plot(kind='barh',alpha=alpha_barchart)
plt.title('Class Distribution')
ax3.set_ylim(-1,len(df['Pclass'].value_counts()))

## figure 4 
# plot kernel density
ax4 = plt.subplot2grid((2,3),(1,0),colspan=2)
df.Age[df.Pclass==1].plot(kind='kde')
df.Age[df.Pclass==2].plot(kind='kde')
df.Age[df.Pclass==3].plot(kind='kde')
plt.xlabel('Age')
plt.legend(('1st class','2nd class','3rd class'),loc='best')
plt.title('Age Distribution within classes')


## figure 5 ###
ax5 = plt.subplot2grid((2,3),(1,2))
df['Embarked'].value_counts().plot(kind='bar',alpha=alpha_barchart)
plt.title('passengers per boarding location')
ax5.set_xlim(-1,len(df['Embarked'].value_counts()))
```

```python
plt.figure(figsize=(18,10))
fig,ax = plt.subplots()
df['Survived'].value_counts().plot(kind='barh',alpha=.65)
ax.set_ylim(-1,len(df['Survived'].value_counts()))
plt.title('1 = Survived , 0 = Died')
```

```python
plt.figure(figsize=(18,6))
df_male = df[df['Sex'] == 'male']['Survived'].value_counts().sort_index()
df_female = df[df['Sex'] == 'female']['Survived'].value_counts().sort_index()

# figure 1 
ax1 = plt.subplot(121) # one row , two column , one number
df_male.plot(kind='barh',label='Male',alpha=.65)
df_female.plot(kind='barh',label='Female',alpha=.65,color='#FA2379')
plt.title('Who Survived ? with respect to Gender')
plt.legend(loc='best')
ax1.set_ylim(-1,2)

ax2= plt.subplot(122)
(df_male / float(df_male.sum())).plot(kind='barh',alpha=.65,label='Male')
(df_female/float(df_female.sum())).plot(kind='barh',alpha=.65,color='#FA2379',label='Female')
plt.title("Who Survived proportionally? with respect to Gender")
plt.legend(loc='best')
ax2.set_ylim(-1,2)
```
