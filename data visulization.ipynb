```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
df.describe()
```

```python
df.drop(['Ticket','Cabin'],axis=1,inplace=True)
```

```python
df
```

```python
# drop NA
df.dropna(inplace=True)
```

# Create graphics

```python
fig = plt.figure(figsize=(18,4),dpi=600)
alpha_barchart = 0.55
alpha = 0.2

## figure 1 ##
ax1 = plt.subplot2grid((2,3),(0,0))  # two row , three column 
df['Survived'].value_counts().plot(kind='bar',alpha=alpha_barchart)
ax1.set_xlim(-1,2)
plt.title('Distribution of Survived ( 1=Survived)')

### figure 2 ##
ax2 = plt.subplot2grid((2,3),(0,1))
plt.scatter(df['Survived'],df['Age'],alpha=alpha)
plt.title('Scatter of Survived by Age (1=Survived) ')
plt.ylabel('Age')
plt.grid(b=True,which='major',axis='y')

## figure 3 
```
